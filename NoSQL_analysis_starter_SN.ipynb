### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20

query = {'scores.Hygiene':20}

# Use count_documents to display the number of documents in the result
hygieneof20 = establishments.count_documents({'scores.Hygiene': 20})
print("the number of establishments with a hygiene score of 20 is", hygieneof20)


# Display the first document in the results using pprint

results = list(establishments.find(query))

pprint(results)
```

```python
# Convert the result to a Pandas DataFrame
df_1 = pd.DataFrame(results)

# Display the number of rows in the DataFrame
length = len(df_1.index)
print("The number of rows in the DataFrame is", length)

# Display the first 10 rows of the DataFrame
df_1.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {'LocalAuthorityName': {'$regex' :'London'} ,'RatingValue' : {'$gte': 4}}

# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print("the number of establishments with London as the Local Authority and has a RatingValue greater than or equal to 4 is", count)


# Display the first document in the results using pprint

london = list(establishments.find(query))

pprint(london)
# Display the first document in the results using pprint
```

```python
# Convert the result to a Pandas DataFrame
df_2= pd.DataFrame(london)

# Display the number of rows in the DataFrame
length = len(df_2.index)
print("The number of rows in the DataFrame is", length)

# Display the first 10 rows of the DataFrame
df_2.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.490142
longitude = 0.08384

query_3 = {'RatingValue' : 5, 
           'geocode.latitude': {'$lte' : latitude + degree_search},
           'geocode.longitude': {'$lte' : longitude + degree_search}}
#sort by the hygiene scores from lowest to greatest
sort_3 =  [('scores.Hygiene', 1)]
#set a limit of 5 establishments
limit_3 = 5

result_3 = list(establishments.find(query_3).sort(sort_3).limit(limit_3))

print ("The number of establishments found is: " + str(len(result_3)))

# Print the results
pprint (result_3)
```

```python
# Convert result to Pandas DataFrame
df_3= pd.DataFrame(result_3)

# Display the number of rows in the DataFrame
length_3 = len(df_3.index)
print("The number of rows in the DataFrame is", length_3)

# Display the first 10 rows of the DataFrame
df_3.head()
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
match_query = {'$match':{'scores.Hygiene': 0}}

# 2. Groups the matches by Local Authority
group_query = {'$group': {'_id': "$LocalAuthorityName", 'count': { '$sum': 1 }}}

# 3. Sorts the matches from highest to lowest
sort_values = {'$sort': { 'count': -1 }}

# Pipeline to group the matches by Local Authority
pipeline = [match_query, group_query, sort_values]

# Print results
results = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print('Number of establishments in each local authority areas with hygiene score of 0 is: ', len(results))

# Print the first 10 results
pprint(results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
df_4= pd.DataFrame(results)

# Display the number of rows in the DataFrame
length_4 = len(df_4.index)
print("The number of rows in the DataFrame is", length_4)

# Display the first 10 rows of the DataFrame
df_4.head(10)
```
